```python
import matplotlib
import pandas as pd
```

```python
# reading the dataset file
dataset=pd.read_csv('car data.csv')
print(dataset.head())
```

```python
# checking for the number of categorical features in the dataset
print(dataset.shape)
```

```python
# checking for the unique values for the fields 'Seller_Type','Transmission' and 'Owner' which are all categorical variable
print(dataset['Seller_Type'].unique())
print(dataset['Transmission'].unique())
print(dataset['Owner'].unique())
```

```python
# checking for any missing or null values
print(dataset.isnull().sum())
```

```python
# checking the details of the dataset
dataset.describe()
```

```python
# printing all the columns that are in the dataset
print(dataset.columns)
```

```python
# by using the 'year' featured created a new feature called the 'numberofyears'
# got the number of years by subtracting the years given in the dataset from the year 2021 in order to know how old a car is or to know for how many years it has been used
# removed the first variable car name because there a huge no of cars and it is not  very good idea to make predictions just based on the car name instead it is better to make predictions based on 'Selling_price and present_price' would be much better
final_dataset=dataset[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven','Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
```

```python
#creating a new variable 'Current_Year' and setting it to 2021
final_dataset['Current_Year']=2021
final_dataset.head()
```

```python
# creating a new feature named 'numberofyears"
final_dataset['numberofyears']=final_dataset['Current_Year']-final_dataset['Year']
print(final_dataset.head())
```

```python
# don't need the fields 'Year' and 'Current_year' so will drop it as my have a feature 'numbderofyears'
# used inplace in order to make sure that the operataion happens in place like a permanent operation
final_dataset.drop(['Year'],axis=1,inplace=True)
print(final_dataset.head())
final_dataset.drop(['Current_Year'],axis=1,inplace=True)
print(final_dataset.head())
```

```python
# trying to convert the categorical features into one hot encoded as there are less number of categories
# get_dummies helps to convert all the things into one hot encoded and then dropping the first column in order to prevent from any dummy variable trap
final_dataset=pd.get_dummies(final_dataset,drop_first=True)
print(final_dataset.head())
print(final_dataset.columns)
```

```python
# finding the correlation
final_dataset.corr()
```

```python
# for displaying the correlation in a diagrametic form I am using seaborn
import seaborn as sns
```

```python
# shows pearplot with respect to each and every feature
print(sns.pairplot(final_dataset))
```

```python
# plotting the above data in the form of heatmap inorder to get better understanding
import matplotlib.pyplot as plt
%matplotlib inline
correlation_matrix=final_dataset.corr()
top_correlation_features=correlation_matrix.index
plt.figure(figsize=(20,20))
```

```python
# plot heatmap
g=sns.heatmap(final_dataset[top_correlation_features].corr(),annot=True,cmap="RdYlGn")
print(g)
print(final_dataset.head())
```

```python
# dependent and independent features
X=final_dataset.iloc[:,1:]
y=final_dataset.iloc[:,0]
print(X.head())
print(y.head())
```

```python
# feature importance
from sklearn.ensemble import ExtraTreesRegressor
models=ExtraTreesRegressor()
print(models.fit(X,y))
print(models.feature_importances_)
```

```python
# plotted graph of feature importances for better visualization
feat_importances=pd.Series(models.feature_importances_, index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print(X_train.shape)
```

```python
# implemnting the random forest regressor
from sklearn.ensemble import RandomForestRegressor
rf_random=RandomForestRegressor()
```

```python
# hyperparameters
import numpy as np
noofestimators=[int(X) for X in np.linspace(start=100,stop=1200,num=12)]
print(noofestimators)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Randomized Search CV
# Number of trees in random forest
# initiated all the parameters with values
noofestimators = [int(X) for X in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
maxnumfeatures = ['auto', 'sqrt']
# Maximum number of levels in tree
maxnumdepth = [int(X) for X in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
minimumnumofsamples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
minimumnumofsamples_leaf = [1, 2, 5, 10]
```

```python
# from all the above given values the best parameter will be automatically selected from the RandomizedSearchCV 
# Create the random grid
# All the parameters should be given in the form of key value pairs
rgrid = {'n_estimators': noofestimators,
               'max_features': maxnumfeatures,
               'max_depth': maxnumdepth,
               'min_samples_split': minimumnumofsamples_split,
               'min_samples_leaf': minimumnumofsamples_leaf}

print(rgrid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
# n_jobs is to consider how many cores need to be used
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = rgrid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
# Doing the prediction
# in order to compare will use the distplot
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions=rf_random.predict(X_test)
```

```python
print(sns.distplot(y_test-predictions))
```

```python
print(plt.scatter(y_test,predictions))
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
